```python
# investifating missing product ingredient list
import requests
from bs4 import BeautifulSoup

# Check a known missing product - COSRX cleanser
url = "https://us.oliveyoung.com/products/UA10124939"
headers = {"User-Agent": "SkinifyBot/1.0"}

r = requests.get(url, headers=headers)
soup = BeautifulSoup(r.text, 'html.parser')
text = soup.get_text(separator="\n")

# Search for ingredient-related text
lines = [l.strip() for l in text.split("\n") if l.strip()]
for i, line in enumerate(lines):
    if any(word in line.lower() for word in ['ingredient', 'aqua', 'water', 'glycerin']):
        print(f"Line {i}: {line[:150]}")
```

```python
def extract_ingredients(lines):
    for i, line in enumerate(lines):
        if "Full Ingredients" in line:
            for j in range(i + 1, min(i + 6, len(lines))):
                candidate = lines[j].strip()
                # Accept comma OR semicolon separated lists
                if len(candidate) > 30 and ("," in candidate or ";" in candidate):
                    # Normalize semicolons to commas for consistency
                    normalized = candidate.replace(";", ",")
                    return normalized
    return None

# Test it on our problem product
test_lines = [l.strip() for l in soup.get_text(separator="\n").split("\n") if l.strip()]
result = extract_ingredients(test_lines)
print(result[:300] if result else "Still not found")
```
