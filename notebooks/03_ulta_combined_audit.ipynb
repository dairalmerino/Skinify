```python
import pandas as pd

# Ulta reviews
ulta = pd.read_csv('../data/ulta/Ulta Skincare Reviews.csv')
print("=== ULTA FILE ===")
print(f"Total rows: {len(ulta)}")
print(f"Columns: {ulta.columns.tolist()}")
print(f"\nSample row:")
print(ulta.iloc[0])

print("\n\n=== COMBINED E-COMMERCE FILE ===")
combined = pd.read_csv('../data/combined/ecommerce_cosmetics.csv', encoding='latin-1')
print(f"Total rows: {len(combined)}")
print(f"Columns: {combined.columns.tolist()}")
print(f"\nSample row:")
print(combined.iloc[0])
```

```python
print("=== WEBSITES ===")
print(combined['website'].value_counts())

print("\n=== COUNTRIES ===")
print(combined['country'].value_counts())

print("\n=== CATEGORIES ===")
print(combined['category'].value_counts())

# Filter to US/Sephora/Ulta only
us_combined = combined[combined['country'].str.lower() == 'usa'] if 'country' in combined.columns else combined
print(f"\nUS products only: {len(us_combined)}")
print(f"Ingredient coverage: {us_combined['ingredients'].notna().sum() / len(us_combined) * 100:.1f}%")
```

```python
# Filter to US Ulta skincare only
ulta_products = combined[
    (combined['country'] == 'USA') & 
    (combined['category'] == 'skincare')
].copy()

print(f"Ulta skincare products: {len(ulta_products)}")
print(f"Products with ingredients: {ulta_products['ingredients'].notna().sum()}")
print(f"Ingredient coverage: {ulta_products['ingredients'].notna().sum() / len(ulta_products) * 100:.1f}%")

print("\n=== SUBCATEGORIES ===")
print(ulta_products['subcategory'].value_counts().to_string())

print("\n=== TOP BRANDS ===")
print(ulta_products['brand'].value_counts().head(15).to_string())
```

```python
ulta_products.to_csv('../data/ulta/ulta_skincare_clean.csv', index=False)
print("Saved to data/ulta/ulta_skincare_clean.csv!")
```
