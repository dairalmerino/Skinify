```python
import pandas as pd

df = pd.read_csv('../data/sephora/product_info.csv')

print(f"Total products: {len(df)}")
print(f"\nColumns: {df.columns.tolist()}")
print(f"\nSample row:")
print(df.iloc[0])
```

```python
# See all primary categories
print("=== PRIMARY CATEGORIES ===")
print(df['primary_category'].value_counts())
```

```python
# Filter to skincare only
sephora_skincare = df[df['primary_category'] == 'Skincare'].copy()

print(f"Skincare products: {len(sephora_skincare)}")
print(f"Products with ingredients: {sephora_skincare['ingredients'].notna().sum()}")
print(f"Missing ingredients: {sephora_skincare['ingredients'].isna().sum()}")
print(f"Ingredient coverage: {sephora_skincare['ingredients'].notna().sum() / len(sephora_skincare) * 100:.1f}%")

print("\n=== SECONDARY CATEGORIES ===")
print(sephora_skincare['secondary_category'].value_counts().to_string())

print("\n=== TOP BRANDS ===")
print(sephora_skincare['brand_name'].value_counts().head(20).to_string())
```

```python
DROP_CATEGORIES = [
    'Value & Gift Sets', 'Mini Size', 'High Tech Tools',
    'Wellness', 'Self Tanners', 'Shop by Concern'
]

sephora_clean = sephora_skincare[
    ~sephora_skincare['secondary_category'].isin(DROP_CATEGORIES)
].copy()

print(f"Before filtering: {len(sephora_skincare)} products")
print(f"After filtering: {len(sephora_clean)} products")
print(f"Products with ingredients: {sephora_clean['ingredients'].notna().sum()}")
print(f"Ingredient coverage: {sephora_clean['ingredients'].notna().sum() / len(sephora_clean) * 100:.1f}%")

# Save
sephora_clean.to_csv('../data/sephora/sephora_skincare_clean.csv', index=False)
print("\nSaved to data/sephora/sephora_skincare_clean.csv!")
```
